In this project done as a part of Data Analyst Nanodegree, we will carry our inquisitive mind into investigating a classic phenomenon from experimental psychology called the Stroop Effect. We will make a hypothesis and then go on statistically to fail to reject it or reject it by looking at a collection of data.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('stroopdata.csv')
df
```

### 1. What is our independent variable? What is our dependent variable?

The **independent variables** are the words condition (congruent or incongruent).

The **dependent variables** are the response time to name the ink colors.

### 2.  What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

H<sub>o</sub> is the **null hypothesis**. It is mathematically defined as $μ_i = μ_c$. It is defined as the mean time of color recognition for congruent words is equal to the mean time for incongruent words.

H<sub>a</sub> is the **alternate hypothesis**. It is mathematically defined as $μ_i ≠ μ_c$. It is defined as the mean time of color recognition for congruent words is not equal to the mean time for incongruent words.

In our example, we have a small sample size and also we don't have any information on the the population standard deviation. It is undoubtedly clear that this experiment has a 2 tailed t-distribution as it's test statistics.

### 5. Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

Let confidence (c) be 90%

$∴\hspace{0.5cm}α = \frac{1-.9}{2} = 0.5$

$∴\hspace{0.5cm}df = n - 1 = 23$

Now,

We reject the Null Hypothesis which states that there is no significant differences between the response time for the two different conditions (Congruent & Incongruent).
